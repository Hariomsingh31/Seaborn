<h1>ðŸŸ¢ 2. Basic Plots
<h2></h2>
    <h2>ðŸ”¹ Histogram

<h1>
<h1>âœ… 1. Histogram (sns.histplot)

<h2>Purpose: Shows the distribution of a single numerical variable.
<h3>
<h2>How it works: Divides data into bins (intervals) and counts how many values fall into each bin.
<h3>
<h2>Use case: Checking spread, skewness, and frequency of values.

```python
data = sns.load_dataset("tips")
sns.histplot(data, x="total_bill", bins=20, kde=True)
plt.show()
```

<h2>
<h2>ðŸ”¹ Scatter Plot
<h2></h2>

<h1>
<h1>âœ… 2. Scatter Plot (sns.scatterplot)
<h3>
<h2>Purpose: Displays the relationship between two numerical variables.
<h3>
<h2>How it works: Each point represents an observation with (x, y) coordinates.
<h3>
<h2>Use case: Detecting correlations, clusters, and outliers.
    <h3>

```python
sns.scatterplot(data,x="total_bill",y="tip",hue="sex")
plt.show()
```

<h1></h1>
<h2>ðŸ”¹ Line Plot
<h2></h2>

<h1>
<h1>âœ… 3. Line Plot (sns.lineplot)
<h3>
<h2>Purpose: Shows how a numerical variable changes over another variable, often time or sequence.
<h3>
<h2>How it works: Connects data points with a line to visualize trends.
<h3>
<h2>Use case: Time series, trends, and continuous data visualization.
    <h3>

```python
sns.lineplot(data,x='total_bill',y='tip',hue='sex',marker='o')
plt.show()
```

<h1>ðŸŸ¢ 3. Categorical Plots
<h2></h2><h2>Categorical plots in Seaborn are used to visualize relationships between a categorical variable and a numerical variable (or between two categorical variables).
They are helpful for comparing groups or categories.</h2>
<h2></h2>

<h1>ðŸ”¹ Bar Plot (sns.barplot)
<h2></h2>
<h2>Purpose: Displays the mean (average) of a numerical variable for each category.
<h3></h3>
<h2>Features: Includes error bars (confidence intervals) by default.
<h3></h3>
<h2>Use case: Comparing average values across groups (e.g., average bill by day).
    <h3></h3>

```python
sns.barplot(data,x="tip",hue='sex' ,)
plt.show()
```

<h1></h1>
<h1>ðŸ”¹ Box Plot (sns.boxplot)
<h2></h2>
<h2>Purpose: Summarizes the distribution of a numerical variable for each category.
<h3></h3>
<h2>How it works: Shows median, quartiles, and outliers using a box-and-whisker design.
<h3></h3>
<h2>Use case: Detecting spread, skewness, and outliers between groups.
    <h3></h3>

```python
sns.boxplot(data,x='size',y='total_bill',hue='smoker')
plt.show()
```

<h1></h1>
<h1>ðŸ”¹ Violin Plot  (sns.violinplot)
<h2></h2>
<h2><b>Purpose:</b> Combines a box plot with a density plot, showing both summary statistics and data distribution.
<h3></h3>
<h2><b>How it works:</b> The width of the violin represents data density.
<h2></h2>
<h2><b>Use case:
</b> Comparing distributions between categories.
    <h3></h3>

```python
sns.violinplot(data,x='day',y='total_bill',hue='sex',)
plt.show()
```

<h1>ðŸ”¹ Strip Plot (sns.stripplot)
<h2>
<h2>Purpose:Plots individual data points along a categorical axis.
<h2></h2>
<h2>How it works: Points may overlap; jitter can spread them out for visibility.
<h2></h2>
<h2>Use case: Seeing the raw distribution of data points.
<h2></h2>

```python
sns.stripplot(data,x='total_bill',y='tip',hue='sex')
plt.show()
```

# ðŸŸ¢ 4. Distribution Plots
<h1>Distribution plots in Seaborn are used to visualize how data values are distributed.
<h1>They help understand patterns like normality, skewness, and spread of numerical variables.

<h2>    
Purpose: Shows the probability density of a variable (smoothed distribution curve).
<h2></h2>
<h2>How it works: Uses Kernel Density Estimation to create a smooth curve instead of discrete bins.
<h2></h2>
<h2>Use case: Understanding the shape of the data distribution.

```python

sns.kdeplot(data,fill=True)
plt.show()
```

<h1>ðŸ”¹ Joint Plot (Scatter + Hist)  (sns.jointplot)
    <h2>
<h2>Purpose: Plots two variables together and shows both scatter and distribution.
<h2>
<h2>How it works: The center plot is a scatter plot, and side plots show distributions of each variable.
<h2>
<h2>Use case: Analyzing relationships + marginal distributions.
<h2></h2>

<h1>ðŸ”¹ Pair Plot (Multiple Variables)  (sns.pairplot)
    <h3></h3>
<h1>Purpose: Shows pairwise relationships between multiple numerical variables.
<h3></h3>
<h2>How it works: Creates a grid of scatter plots for every combination of variables.
<h3></h3>
<h2>Use case: Exploratory Data Analysis (EDA) for multivariate datasets.
    <h3></h3>

```python
df=pd.DataFrame(data)

sns.pairplot(df)
plt.show()
```
